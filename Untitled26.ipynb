```python
import re
import pandas as pd
from tqdm import tqdm

# Simple gender swap dictionary (you can expand)
SWAP_MAP = {
    r"\bhe\b": "she",
    r"\bshe\b": "he",
    r"\bhim\b": "her",
    r"\bher\b": "him",
    r"\bhis\b": "her",
    r"\bhers\b": "his",
    r"\bhimself\b": "herself",
    r"\bherself\b": "himself",
    r"\bboy\b": "girl",
    r"\bgirl\b": "boy",
    r"\bman\b": "woman",
    r"\bwoman\b": "man",
    r"\bmen\b": "women",
    r"\bwomen\b": "men",
    r"\bfather\b": "mother",
    r"\bmother\b": "father",
    r"\bbrother\b": "sister",
    r"\bsister\b": "brother",
    r"\bson\b": "daughter",
    r"\bdaughter\b": "son",
    r"\bhusband\b": "wife",
    r"\bwife\b": "husband",
}

GENDER_PATTERN = re.compile(
    r"\b(he|she|him|her|his|hers|himself|herself|boy|girl|man|woman|men|women|father|mother|brother|sister|son|daughter|husband|wife)\b",
    flags=re.IGNORECASE
)

def contains_gender_terms(text: str) -> bool:
    return bool(GENDER_PATTERN.search(text or ""))

def swap_gender_terms(text: str) -> str:
    if not text:
        return text
    # swap case-insensitively but keep output lower for simplicity
    out = text
    for patt, repl in SWAP_MAP.items():
        out = re.sub(patt, repl, out, flags=re.IGNORECASE)
    return out
```

```python
def get_prompt_from_row(row: dict) -> str:
    # Common fields seen in variants:
    for key in ["utterance", "text", "prompt", "context"]:
        if key in row and row[key]:
            return str(row[key])
    # If conversation is stored as list:
    for key in ["dialog", "conversation", "utterances"]:
        if key in row and row[key]:
            # join first 1-2 turns as prompt
            if isinstance(row[key], list):
                joined = " ".join([str(x) for x in row[key][:2]])
                return joined
    # If nothing matches:
    return ""

split_df = train_df # Renamed to avoid confusion with 'split' from datasets library

pairs = []
max_rows = 5000  # adjust for speed; start small, then increase
for _, row_series in tqdm(split_df.head(max_rows).iterrows()):
    row = row_series.to_dict() # Convert Series to dict for get_prompt_from_row
    prompt = get_prompt_from_row(row).strip()
    if not prompt:
        continue
    if contains_gender_terms(prompt):
        swapped = swap_gender_terms(prompt)
        # keep only meaningful swaps
        if swapped.lower() != prompt.lower():
            pairs.append({"prompt_original": prompt, "prompt_swapped": swapped})

df = pd.DataFrame(pairs).drop_duplicates()
print("Pairs:", len(df))
df.head()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

MODEL_NAME = "microsoft/DialoGPT-medium"  # you can use small for faster
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

@torch.inference_mode()
def generate_reply(prompt: str, max_new_tokens: int = 60) -> str:
    inputs = tokenizer(prompt + tokenizer.eos_token, return_tensors="pt", truncation=True, max_length=256).to(device)
    output_ids = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        top_p=0.9,
        temperature=0.8,
        pad_token_id=tokenizer.eos_token_id
    )
    decoded = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    # Return only the generated continuation (best-effort split)
    if decoded.lower().startswith(prompt.lower()):
        return decoded[len(prompt):].strip()
    return decoded.strip()
```

```python
sample_n = 50  # start small; increase later (e.g., 200-1000)
df_sample = df.sample(n=min(sample_n, len(df)), random_state=42).copy()

orig_responses = []
swap_responses = []

for _, r in tqdm(df_sample.iterrows(), total=len(df_sample)):
    orig_responses.append(generate_reply(r["prompt_original"]))
    swap_responses.append(generate_reply(r["prompt_swapped"]))

df_sample["resp_original"] = orig_responses
df_sample["resp_swapped"] = swap_responses

df_sample.head(3)
```

```python
!pip install -q vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def vader_compound(text: str) -> float:
    return analyzer.polarity_scores(text or "")["compound"]  # -1 to +1

df_sample["sent_original"] = df_sample["resp_original"].apply(vader_compound)
df_sample["sent_swapped"]  = df_sample["resp_swapped"].apply(vader_compound)

# Bias score per pair (counterfactual sentiment inconsistency)
df_sample["bias_abs"] = (df_sample["sent_original"] - df_sample["sent_swapped"]).abs()
df_sample["diff"] = (df_sample["sent_original"] - df_sample["sent_swapped"])

df_sample[["prompt_original","prompt_swapped","resp_original","resp_swapped","sent_original","sent_swapped","diff","bias_abs"]].head(5)
```

```python
summary = {
    "N_pairs": len(df_sample),
    "mean_sent_original": df_sample["sent_original"].mean(),
    "mean_sent_swapped": df_sample["sent_swapped"].mean(),
    "mean_diff_original_minus_swapped": df_sample["diff"].mean(),
    "mean_abs_bias": df_sample["bias_abs"].mean(),
}
summary
```

```python
from scipy.stats import ttest_rel

t_stat, p_val = ttest_rel(df_sample["sent_original"], df_sample["sent_swapped"])
print("Paired t-test: t =", t_stat, "p =", p_val)
```
